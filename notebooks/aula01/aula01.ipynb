```python
def find(lista):
    alerta = []
    for i in range(len(lista)):
        if lista[i] > 100:
            alerta.append(lista[i])
    if len(alerta) == 0:
        print("Sistema seguro")
    else:
        print("Revisar valores: ", alerta)

valores_brutos = [10, 50, 150, 20, 300]
find(valores_brutos)
```

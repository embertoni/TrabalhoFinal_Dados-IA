```python
df['Primary_Use_Case'] = df['Primary_Use_Case'].map({'Copywriting/Drafting':0, 'Ideation':1, 'Summarizing_Reading':2, 'Debugging/Troubleshooting':3,'Direct_Answer_Generation':4 }).astype('category')
df
```

```python
df.info(memory_usage="deep")
df.memory_usage(deep=True)
```

```python
df["Major_Category"].unique()
df['Major_Category'] = df['Major_Category'].map({'Humanities':0, 'Medical':1, 'Business':2, 'STEM':3,'Arts':4 }).astype('category')
df
```

```python
df["Year_of_Study"].unique()
df['Year_of_Study'] = df['Year_of_Study'].map({'Senior':0, 'Junior':1, 'Freshman':2, 'Sophomore':3,'Graduate':4 }).astype('category')
df
```

```python
df["Prompt_Engineering_Skill"].unique()
df['Prompt_Engineering_Skill'] = df['Prompt_Engineering_Skill'].map({'Beginner':0, 'Advanced':1, 'Intermediate':2}).astype('category')
df
```

```python
df["Institutional_Policy"].unique()
df['Institutional_Policy'] = df['Institutional_Policy'].map({'Allowed_With_Citation':0, 'Strict_Ban':1, 'Actively_Encouraged':2}).astype('category')
df
```

```python
df["Burnout_Risk_Level"].unique()
df['Burnout_Risk_Level'] = df['Burnout_Risk_Level'].map({'High':0, 'Low':1, 'Medium':2}).astype('category')
df
```

```python
df.info(memory_usage="deep")
df.memory_usage(deep=True)
```

```python
df["Anxiety_Level_During_Exams"] = df["Anxiety_Level_During_Exams"].astype("int32")
```

```python
df["Post_Semester_GPA"] = df["Post_Semester_GPA"].astype("float32")
```
